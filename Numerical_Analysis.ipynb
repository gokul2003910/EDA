Numerical Analysis

Correlation it ranges from -1 to +1
  +1 -> positive correlation
  -1 -> negative correlation
  0  -> no correlation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
%matplotlib inline
```

```python
telco_base_data = pd.read_csv('Churn_Modelling.csv')
```

```python
telco_base_data.select_dtypes(include=[np.number]).corr()
```

```python
plt.figure(figsize=(10, 8))
telco_base_data.select_dtypes(include=[np.number]).corr()['Exited'].sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(telco_base_data.select_dtypes(include=[np.number]).corr(), cmap='Paired')
```

```python
telco_base_data['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
telco_base_data1 = telco_base_data.loc[telco_base_data['Exited'] == 1]
```

```python
telco_base_data1['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
Tot = sns.kdeplot(telco_base_data.Age[(telco_base_data["Exited"]==0)],
                  color="Red", shade=True)
Tot = sns.kdeplot(telco_base_data.Age[(telco_base_data["Exited"]==1)],
                  color="Blue", shade=True)
Tot.legend(["No Churn", "Churn"], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Age')
Tot.set_title('Age by churn')
```

```python
Tot = sns.kdeplot(telco_base_data.Tenure[(telco_base_data["Exited"]==0)],
                  color="Red", shade=True)
Tot = sns.kdeplot(telco_base_data.Tenure[(telco_base_data["Exited"]==1)],
                  color="Blue", shade=True)
Tot.legend(["No Churn", "Churn"], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Tenure')
Tot.set_title('Tenure by churn')
```
